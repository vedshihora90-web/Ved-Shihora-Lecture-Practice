# series in python 

#### a series is a one dimantional labeled array capeble of holding any data types

```python
import pandas as pd
```

```python
data=pd.Series([])
print(data)
data=pd.Series([1,2,3,4,5])
print(data)
```

#### series with custom index

```python
data=pd.Series(
                [56,89,87],
                index=["maths","english","science"])
print(data)
```

#### access series data

```python
print(data["maths"])
```

# data frames in python

#### a data farme is a two dimensional labeled tubular structure considering of rows and colums

```python
student={"name":["ved","shihora","sachin"],
         "marks":[56,89,87],
         "city":["pune","mumbai","delhi"]}
df=pd.DataFrame(student)    
print(df)
```

### creation of data frames

```python
students=[["ved",56,"pune"],
          ["shihora",89,"mumbai"],
          ["sachin",87,"delhi"]]
df=pd.DataFrame(students,columns=["name","marks","city"])       
print(df)
```

### exploring data frames

```python
student={"name":["ved","shihora","sachin"],
         "marks":[56,89,87],
         "city":["pune","mumbai","delhi"]}
df=pd.DataFrame(student)
print(df)
```

#### display last row

```python
print(df.tail(2))
print(df.tail(1))
print(df.tail(3))
print(df.head(2))
```

#### shape of data frames

```python
print(df.shape)
print(df.columns)
```

#### statistical info

```python
print(df.describe())
print(df.info())
```

### operation on data frames

#### select single colum

```python
print(df["name"])
```

#### select multiple colum

```python
print(df[["name","marks"]])
```

#### add new colum

```python
df["grade"]=["A","B","A"]
print(df)
```

#### update colum

```python
df["marks"]=df["marks"]+10
print(df)
```

#### delete colum

```python
df.drop("grade",axis=1,inplace=True)
print(df)
```

#### filter rows

```python
r=df[df["marks"]>80]
print(r)
```

#### sorting data
1. asecending

2. discending

```python
print(df.sort_values("marks"))
print(df.sort_values("marks",ascending=False))
print(df["marks"].mean())
print(df["marks"].max())
print(df["marks"].min())
print(df["marks"].sum())
```

# pandas series all types with nan values

#### what is nan?

#### nan is not a number repersent missing or undefined values in pandas

```python
import numpy as np
s=pd.Series([1,2,3,np.nan,5])
print(s)
```

#### string series with nan

```python
s=pd.Series(['amit','vijay','sachin',np.nan,'rohit'])
print(s)
```

#### series with nan

```python
s=pd.Series(['amit',10,None,np.nan,10.4])
print(s)
print(s.isnull())
print(s.notnull())
print(s.isnull().sum())
```

### fill nan values

#### using fillna()

```python
s=pd.Series([10,None,20,30])
s=s.fillna(0)
print(s)
```

#### using dropna()

```python
s=pd.Series(['amit','vijay','sachin',np.nan,'rohit'])
print(s.dropna())
```
