### change data types

```python
import pandas as pd

print("=========Read First Sheet=========")
df = pd.read_excel('student_data.xlsx')
print(df)
```

```python
print(df.dtypes)
df["age"] = df["age"].astype(int)
print(df.dtypes)
```

### date & time

```python
df["joining_date"] = pd.to_datetime(df["joining date"], dayfirst=True)
df["year"] = df["joining_date"].dt.year
df["month"] = df["joining_date"].dt.month_name()
df["day"] = df["joining_date"].dt.day_name()
print(df[["joining_date", "year", "month", "day"]])
```

#### filter data

```python
print("\n ========marks.80========")
r=df[df["marks"] > 80]
print(r)
city=df[df["city"]=="surat"]
print(city)
```

#### sorting data

```python
print("\n=========ascending==========")
r=df.sort_values(by="marks", ascending=True)
print(r)
```

#### add new column

```python
df["r"]=df["marks"].apply(lambda x: "pass" if x>35 else "fail")
print(df)
```

#### rename column

```python
df.rename(columns={"joining date":"joining_date"}, inplace=True)
print(df.columns)
print(df)
```

#### delete column

```python
df.drop("age", axis=1, inplace=True)
print(df)
```

#### save updated data into excel

```python
df.to_excel("student_data.xlsx", sheet_name="Sheet1", index=False)
print("\n updated data saved to student_data.xlsx")
```

#### pandas shorting,filtering,aggregation,merge and join

```python
print("="*90)
import pandas as pd
student_data = {
    "id": [1, 2, 3, 4, 5],
    "name": ["Alice", "Bob", "Charlie", "David", "Eva"],    
    "age": [20, 21, 19, 22, 20],
    "city": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"],
    "marks": [85, 90, 78, 92, 88],
    "course": ["Math", "Science", "History", "Math", "Science"]
}
df = pd.DataFrame(student_data)
print(df)
```

### merge data

```python
print("\n average marks by course")
print(df.groupby("course")["marks"].mean())
print("\n maximum marks by course")
print(df.groupby("course")["marks"].max())
print("\n minimum marks by course")
print(df.groupby("course")["marks"].min())
print("\n count of students by course")
print(df.groupby("course")["id"].count())
```

```python
fees={
    "id": [1, 2, 3, 4, 5],
    "fees_paid": [1000, 1500, 1200, 1800, 2000]

}
fees_df=pd.DataFrame(fees)
print("\n fees data")
print(fees_df)
```

### mergw

```python
print("inner message")
inner_join = pd.merge(df, fees_df, on="id", how="inner")
print("\n inner join result")
print(inner_join)
```
