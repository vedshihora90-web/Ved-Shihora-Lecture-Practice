# create, read, and write csv, excel & json

```python
import pandas as pd
```

### create dictonary

```python
student_data={"name":["vivek","vraj","nihar","jahi","nidhi"],
              "age":[20,22,21,23,20],
              "grade":["A","B","A","C","B"],
              "marks":[89,78,67,87,56],
              "city":["ahemadabad","surat","vadodara","bhavanagar","jamnagar"],
              "joining date":["20-06-2025","21-06-2025","22-06-2025","23-06-2025","24-06-2025"]}
df=pd.DataFrame(student_data)
print(df)
```

### to_csv

```python
df.to_csv("student_data.csv",index=False)
print("\n CSV File created successfully.")
```

### to_excel

```python
df.to_excel("student_data.xlsx",index=False)
print("Excel file created successfuly!!")
```

### json

```python
df.to_json("student_data.json",orient="records",index=4)
print("json file created successfully!!")
```

### read csv file

```python
csv_data=pd.read_csv("student_data.csv")
print(csv_data)
```
