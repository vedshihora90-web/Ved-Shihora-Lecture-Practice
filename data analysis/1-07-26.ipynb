## prectical tasks for pivoting, re-indexing, groupby and transform in panadas


```python
import pandas as pd
print("="*40)
print("1.pivoting data frames")
print("="*40)
```

### monthly sales pivot

```python
sales=pd.DataFrame({
    "month":["January","January","February","February"],
    "property":["A","B","A","B"],
    "sales":[100,200,300,400]

})
print(sales)
pivot_sales=sales.pivot(index="month",columns="property",values="sales")
month_order=["January","February"]
pivot_sales=pivot_sales.reindex(month_order)
print(pivot_sales)
```

## students marks pivot

```python
marks=pd.DataFrame({
    "student":["A","A","B","B"],
    "subject":["Maths","Science","Maths","Science"],
    "marks":[90,80,70,60]
})
print(marks)
pivot_marks=marks.pivot(index="student",columns="subject",values="marks")
print(pivot_marks)
```

```python
employee=pd.DataFrame({
    "employee":["A","A","B","B"],
    "department":["HR","IT","HR","IT"],
    "hours":[10,20,30,40]
})


pivot_employee=employee.pivot_table(
                        index="department",
                         columns="employee", 
                        values="hours",
                        aggfunc="sum"
)

print(pivot_employee)
```

### fruits price pivot

```python
fruits=pd.DataFrame({
    "fruit":["Apple","Apple","Banana","Banana"],
    "region":["North","South","North","South"],
    "sales":[100,200,300,400]
})
pivot_fruits=fruits.pivot_table(
    index="region",
    columns="fruit",
    values="sales",
    aggfunc="sum"
)
print(pivot_fruits)
```

### shipment  delivery time pivot

```python
shipment=pd.DataFrame({
    "city":["New York","New York","Los Angeles","Los Angeles"],
    "carrier":["FedEx","UPS","FedEx","UPS"],
    "days":[2,3,4,5]
})
pivot_shipment=shipment.pivot_table(
    index="city",
    columns="carrier",
    values="days",
    aggfunc="mean"
)
print(pivot_shipment)
```

### re-index employee ids

```python
employee=pd.DataFrame({
    "employee":["A","A","B","B"],
    "salary":[1000,2000,3000,4000],
})
employee.index=[1001,1002,1003,1004]
print(employee)
```

#### total sales by region

```python
sales=pd.DataFrame({
    "region":["North","North","South","South"],
    "product":["A","B","A","B"],
    "sales":[100,200,300,400]
})
region_sales=sales.groupby("region")["sales"].sum()
print(region_sales)
```
