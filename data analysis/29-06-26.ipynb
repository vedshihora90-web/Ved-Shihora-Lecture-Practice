# pandas read and manipulate one or multiple excel sheet

```python
import pandas as pd
```

```python
print("=========Read First Sheet=========")
df=pd.read_excel('student_data.xlsx')
print(df)
```

#### read specific sheet

```python
sheet2=pd.read_excel("student_data.xlsx",sheet_name="Sheet2")
print(sheet2)
```


#### select data

#### read sheet by indax

```python
print("======= read second sheet=========")
sheet2=pd.read_excel("student_data.xlsx",sheet_name=1)
print(sheet2)
```

#### read multiple sheets

```python
print("=========read multiple sheet ============")
multiple_sheet=pd.read_excel("student_data.xlsx",sheet_name=[0,1])
print(multiple_sheet)
```

#### read all sheets

```python
print("=========read all sheet========")
all_sheet=pd.read_excel("student_data.xlsx",sheet_name=None)
for sheet_name,data in all_sheet.items():
    print('----------------------')
    print(data)
```

#### missing values

```python
print("============mising vcalues===============")
print(df.isnull().sum())
df["marks"]=df["marks"].fillna(df["marks"].mean())
df["city"]=df["city"].fillna('unknown')
print(df)
```

#### remove duplicate recored

```python
print("\n duplicate rows:")
df.duplicated().sum()
df=df.drop_duplicates()
print("\n after duplicate:")
print (df)
```
